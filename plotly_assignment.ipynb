Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset

```python
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset using seaborn
titanic = sns.load_dataset("titanic")

# Create a scatter plot using Plotly express
scatter_plot = px.scatter(titanic, x="age", y="fare", title="Scatter Plot of Age vs Fare in Titanic")

# Show the plot
scatter_plot.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
histogram = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    facet_col="smoker",
    pattern_shape="day",
    title="Histogram of Total Bill Amount by Sex, Smoker, and Day",
    labels={"sex": "Sex", "total_bill": "Total Bill Amount", "day": "Day", "smoker": "Smoker"}
)

# Show the plot
histogram.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
import plotly.express as px
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
scatter_matrix = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset"
)

# Show the plot
scatter_matrix.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

```python
import plotly.express as px
import numpy as np

# Generating some random data for demonstration purposes
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Creating a distplot using Plotly Express
fig = px.histogram(data, nbins=30, title='Distribution Plot', labels={'value': 'Data'})

# Show the plot
fig.show()
```
